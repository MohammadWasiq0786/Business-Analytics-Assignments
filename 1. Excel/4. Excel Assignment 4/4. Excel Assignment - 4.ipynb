**Ans :** String formatting in Excel is performed using the CONCATENATE function or the ampersand operator (&). The CONCATENATE function is used to join two or more strings together, while the ampersand operator can be used to concatenate strings and values.

**Here are some examples of how to perform string formatting in Excel :**

1. Using the CONCATENATE function:

Suppose we have the following data in cells A1 and B1:

A1: "John"

B1: "Doe"

To concatenate these two strings, you can use the CONCATENATE function as follows:

`=CONCATENATE(A1, " ", B1)`

This will produce the output "John Doe", with a space between the two strings.

2. Using the ampersand operator:

We can achieve the same result using the ampersand operator as follows:

`=A1 & " " & B1`

This will also produce the output "John Doe".

3. Concatenating strings with values:

Suppose we have the following data in cells A1 and B1:

A1: "John"

B1: 25

To concatenate the string in A1 with the value in B1, you can use the ampersand operator as follows:

`=A1 & " is " & B1 & " years old"`

This will produce the output "John is 25 years old".

4. Using cell references:

We can also use cell references to concatenate strings in Excel. Suppose you have the following data in cells A1 and B1:

A1: "John"

B1: "Doe"

C1: "Full Name:"

To concatenate the strings in A1 and B1 with the string in C1, you can use the ampersand operator as follows:

`=C1 & " " & A1 & " " & B1`

This will produce the output **"Full Name: John Doe"**.

**Q5 :** Create reports to generate mark sheets of students in excel where percentages and addition of marks should be done using formulas.

**Ans :** Follow the below steps:

1. Create a table with the following headers in cells A1 to E1:

| Student Name | Subject | Marks |
|--------------|---------|-------|

2. Enter the Roll Number and Name of the students in cells A2 and B2 respectively.
3. Enter the marks obtained by each student in Subject 1 and Subject 2 in cells C2 and D2 respectively.

| Student Name | Subject | Marks |
|--------------|---------|-------|
| John         | Math    | 85    |
| John         | Science | 90    |
| Jane         | Math    | 80    |
| Jane         | Science | 95    |
| Bob          | Math    | 90    |
| Bob          | Science | 85    |
| Mary         | Math    | 95    |
| Mary         | Science | 80    |

4. To calculate the Total Marks obtained by each student, enter the following formula in cell E2:

`=SUM(C2:D2)`

This will add the marks obtained in Subject 1 and Subject 2 to give the total marks.

5. To calculate the percentage of marks obtained by each student, enter the following formula in cell F2:

`=(E2/200)*100`

Assuming that the total marks for each subject is 100, we divide the total marks obtained by 200 (100 + 100) and multiply it by 100 to get the percentage.

6. Copy the formulas in cells E2 and F2 and paste them in the respective columns for all the other students.

7. To format the percentages to show two decimal places, select the column containing the percentages (column F), right-click, and select "Format Cells". In the "Number" tab, select "Percentage" and set the decimal places to 2.
